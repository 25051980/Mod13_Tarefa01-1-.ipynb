3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 
    

```python
# Preparar a fórmula para o modelo
# Inclua todas as variáveis, transformando variáveis categóricas em dummies
# As variáveis categóricas são: 'sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'
formula = 'np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + educacao + estado_civil + tipo_residencia + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia'

# Criar matrizes de design usando Patsy
y, X = dmatrices(formula, df, return_type='dataframe')

# Ajustar modelo
modelo = sm.OLS(y, X).fit()

# Ver resumo do modelo
print(modelo.summary())
```

```python
# Identificar a categoria mais frequente para cada variável categórica
categorias_referencia = {
    'sexo': df['sexo'].value_counts().idxmax(),
    'tipo_renda': df['tipo_renda'].value_counts().idxmax(),
    'educacao': df['educacao'].value_counts().idxmax(),
    'estado_civil': df['estado_civil'].value_counts().idxmax(),
    'tipo_residencia': df['tipo_residencia'].value_counts().idxmax()
}

# Preparar a fórmula para o modelo, definindo a categoria mais frequente como referência
formula = f'np.log(renda) ~ C(sexo, Treatment(categorias_referencia["sexo"])) + posse_de_veiculo + posse_de_imovel + C(tipo_renda, Treatment(categorias_referencia["tipo_renda"])) + C(educacao, Treatment(categorias_referencia["educacao"])) + C(estado_civil, Treatment(categorias_referencia["estado_civil"])) + C(tipo_residencia, Treatment(categorias_referencia["tipo_residencia"])) + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia'

# Criar matrizes de design usando Patsy
y, X = dmatrices(formula, df, return_type='dataframe')

# Ajustar modelo
modelo = sm.OLS(y, X).fit()

# Ver resumo do modelo
print(modelo.summary())
```

```python
# Preparar a fórmula para o modelo
formula = 'np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + educacao + estado_civil + tipo_residencia + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia'

# Criar matrizes de design usando Patsy
y, X = dmatrices(formula, df, return_type='dataframe')

# Ajustar modelo
modelo = sm.OLS(y, X).fit()

# Ver resumo do modelo
print(modelo.summary())

# Avaliar os parâmetros do modelo
params = modelo.params
pvalues = modelo.pvalues
conf = modelo.conf_int()

for param in params.index:
    print(f"\nCoeficiente para {param}:")
    print(f" - Valor: {params[param]}")
    print(f" - p-value: {pvalues[param]}")
    print(f" - Intervalo de confiança: {conf.loc[param]}")

    # Verifica a significância
    if pvalues[param] < 0.05:
        print(" - Significativo estatisticamente")
    else:
        print(" - Não significativo estatisticamente")
```

```python
# Fórmula inicial
formula = 'np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + educacao + estado_civil + tipo_residencia + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia'

# Ajustar modelo inicial
y, X = dmatrices(formula, df, return_type='dataframe')
modelo_inicial = sm.OLS(y, X).fit()

# Identificar a variável menos significativa
pvalues = modelo_inicial.pvalues
variavel_menos_significativa = pvalues.idxmax()  # Obtém o índice do maior p-value

# Remover a variável menos significativa e ajustar um novo modelo
formula_nova = formula.replace(f'+ {variavel_menos_significativa}', '')  # Remove a variável da fórmula
y, X = dmatrices(formula_nova, df, return_type='dataframe')
modelo_novo = sm.OLS(y, X).fit()

# Comparar os modelos
print("Modelo Inicial:\n", modelo_inicial.summary())
print("\nModelo Novo:\n", modelo_novo.summary())

# Avalie os indicadores como R² ajustado, AIC, BIC e p-values
```

```python
# Continuação do código anterior...

# Comparar coeficientes entre os dois modelos
print("\nComparação dos Coeficientes:")
for param in modelo_inicial.params.index:
    if param in modelo_novo.params:
        print(f"{param}:")
        print(f" - Modelo Inicial: {modelo_inicial.params[param]}, p-value: {modelo_inicial.pvalues[param]}")
        print(f" - Modelo Novo: {modelo_novo.params[param]}, p-value: {modelo_novo.pvalues[param]}")
        print(f" - Diferença: {modelo_novo.params[param] - modelo_inicial.params[param]}")
    else:
        print(f"{param} foi removido no modelo novo.")

# Avaliar mudanças significativas nos coeficientes
```

```python
# Fórmula inicial
formula = 'np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + educacao + estado_civil + tipo_residencia + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia'

# Função para ajustar o modelo e retornar o modelo e a fórmula atualizados
def ajustar_modelo(formula, df):
    y, X = dmatrices(formula, df, return_type='dataframe')
    modelo = sm.OLS(y, X).fit()
    return modelo, formula

# Ajustar o modelo inicial
modelo, formula_atual = ajustar_modelo(formula, df)
print("Modelo Inicial:\n", modelo.summary())

# Remover iterativamente as variáveis menos significativas
while True:
    pvalues = modelo.pvalues
    pvalues = pvalues.drop('Intercept')  # Ignorar o intercepto
    max_pvalue = pvalues.max()
    if max_pvalue < 0.05:
        break  # Parar se todos os p-values forem menores que 5%
    variavel_a_remover = pvalues.idxmax()  # Obter a variável com o maior p-value
    formula_atual = formula_atual.replace(f'+ {variavel_a_remover}', '')  # Remover variável
    modelo, formula_atual = ajustar_modelo(formula_atual, df)

# Comparar o modelo final com o inicial
print("\nModelo Final:\n", modelo.summary())

# Avaliar mudanças nos indicadores como R² ajustado, AIC, BIC
```
